```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import optimize

#3.1
def f(x1,x2):
  return (x1**4) - (2 * x2 * x1**2) + x2**2 + x1**2 - 2*x1 + 5

x1 = np.linspace(-20, 20, 20)
x2 = np.linspace(-20, 20, 20)
X1,X2 = np.meshgrid(x1,x2)

plt.figure(figsize=(5,3))
plt.title('Contour of the given function')
plt.contour(X1,X2,f(X1,X2))
plt.show()

#3.2
def f(x):
  return (x[0]**4) - (2 * x[1] * x[0]**2) + x[1]**2 + x[0]**2 - 2*x[0] + 5

#initail guesses
n = 2
x0 = np.zeros(n)
x0[0] = 0.25
x0[1] = 2

#optimize
b1 = (-1,5)
b2 = (0,3)
bnds = (b1,b2)

solution = optimize.minimize(f, x0, method = 'SLSQP', bounds = bnds)
x = solution.x

#show final objection
print('Final objection: ' + str(f(x)))

#print solution
print('x1 = ' + str(x[0]))
print('x2 = ' + str(x[1]))
```
