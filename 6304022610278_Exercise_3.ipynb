```python
p = figure(title = 'Histogram of Arrival Delay', x_range=(-70, 130))
p.xaxis.axis_label = "Delay (min)]"
p.yaxis.axis_label = "Number of Flights"

#histogram
df = df[~np.isnan(df['arr_delay'])]
hist, edges = np.histogram(df['arr_delay'], bins=range(-60, 120, 5))
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], line_color="black", fill_color="red", alpha=0.8)
show(p)
```
