##Import all the necessery libraries

```python
# import KMeans
from sklearn.cluster import KMeans

# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

##Download the dataset and turn it to numpy array

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/milaan9/Clustering-Datasets/master/02.%20Synthetic/2spiral.csv',header=0)
# points = dataset[0]
points = dataset[['a0','a1']].to_numpy()
```

##Implement the K-Means Clustering

```python
# create kmeans object
kmeans = KMeans(n_clusters=4)
# fit kmeans object to data
kmeans.fit(points)
# print location of clusters learned by kmeans object
print(kmeans.cluster_centers_)
# save new clusters for chart
y_km = kmeans.fit_predict(points)
```

##Visualize the data

```python
plt.scatter(points[y_km ==0,0], points[y_km == 0,1], s=100, c='red')
plt.scatter(points[y_km ==1,0], points[y_km == 1,1], s=100, c='black')
plt.scatter(points[y_km ==2,0], points[y_km == 2,1], s=100, c='blue')
plt.scatter(points[y_km ==3,0], points[y_km == 3,1], s=100, c='cyan')
```

##Implement the Agglomerative Hierarchical Clustering

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(points, method='ward'))
# create clusters
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'ward')
# save clusters for chart
y_hc = hc.fit_predict(points)
```

##Visualize the data

```python
plt.scatter(points[y_hc ==0,0], points[y_hc == 0,1], s=100, c='red')
plt.scatter(points[y_hc==1,0], points[y_hc == 1,1], s=100, c='black')
plt.scatter(points[y_hc ==2,0], points[y_hc == 2,1], s=100, c='blue')
plt.scatter(points[y_hc ==3,0], points[y_hc == 3,1], s=100, c='cyan')
```

##Comparison

```python
fig, axs = plt.subplots(1, 2,figsize=(15,5))
fig.suptitle('k-mean clustering vs agglomerative hierarchical clustering',fontsize=20)
axs[0].scatter(points[y_km ==0,0], points[y_km == 0,1], s=100, c='red')
axs[0].scatter(points[y_km ==1,0], points[y_km == 1,1], s=100, c='black')
axs[0].scatter(points[y_km ==2,0], points[y_km == 2,1], s=100, c='blue')
axs[0].scatter(points[y_km ==3,0], points[y_km == 3,1], s=100, c='cyan')
axs[0].set_title("k-mean clustering",fontsize=10)

axs[1].scatter(points[y_hc ==0,0], points[y_hc == 0,1], s=100, c='red')
axs[1].scatter(points[y_hc==1,0], points[y_hc == 1,1], s=100, c='black')
axs[1].scatter(points[y_hc ==2,0], points[y_hc == 2,1], s=100, c='blue')
axs[1].scatter(points[y_hc ==3,0], points[y_hc == 3,1], s=100, c='cyan')
axs[1].set_title('hierarchical clustering',fontsize=10)
```
