```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#setup
prob = 1/101
xs = np.linspace(0,1,num = 101)
prior = pd.Series(prob, xs)
likelihood_blue = xs
likelihood_yellow = 1-xs
```

##one blue ball

```python
unnorm = prior * likelihood_blue
posterior = unnorm / unnorm.sum()
posterior.plot()
plt.xlabel('Fraction of blue balls')
plt.ylabel('Probability')
plt.title('Posterior, one blue ball');
```

##two blue ball

```python
prior = posterior
unnorm = prior * likelihood_blue
posterior = unnorm / unnorm.sum()

posterior.plot()
plt.xlabel('Fraction of blue balls')
plt.ylabel('Probability')
plt.title('Posterior, two blue ball');
```

##three blue ball

```python
prior = posterior
unnorm = prior * likelihood_blue
posterior = unnorm / unnorm.sum()

posterior.plot()
plt.xlabel('Fraction of blue balls')
plt.ylabel('Probability')
plt.title('Posterior, three blue ball');
```
