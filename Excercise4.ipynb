#1-2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/sapichat/ML_Data_Shared/main/data_age_salary.txt"
dataset = pd.read_csv(url,sep="\t")

#print(dataset)

age = dataset.iloc[:,0]
salary = dataset.iloc[:,1]
# print(age)
# print(salary)
plt.scatter(age,salary,c='red')
plt.grid()
plt.show()
```

#3

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
x = np.array(age).reshape((-1,1))
y = np.array(salary)

model.fit(x,y)
y_pred = model.predict(x)
plt.scatter(age,salary,c='red')
plt.plot(x,y_pred,linewidth=3)
plt.grid()
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

#4

```python
r_sq = model.score(x,y)
print(f"coefficient of determination: {r_sq}")
```

# 5

```python
from sklearn.metrics import mean_squared_error

url_test = "https://raw.githubusercontent.com/sapichat/ML_Data_Shared/main/data_age_salary_forTest.txt"
dataset_test = pd.read_csv(url_test,sep="\t")
age_test = dataset_test.iloc[:,0]
salary_test = dataset_test.iloc[:,1]

x_test = np.array(age_test).reshape((-1,1))
y_test = np.array(salary_test)

y_test_pred = model.predict(x_test)
```

#6

```python
print(mean_squared_error(y_test,y_test_pred))
```

# 7

```python
plt.scatter(age_test,salary_test,c='red')
plt.plot(x_test,y_test_pred)
plt.grid()
plt.title("Age vs Salary--TEST")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
