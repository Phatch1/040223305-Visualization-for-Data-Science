```python
from scipy import optimize as opt
import numpy as np

def f(x):
  return (x[0]**4)-(2*x[1]*(x[0]**2))+x[1]**2+x[0]**2-(2*x[0])+5

def constraint1(x):
  return -1*(x[0]+0.25)**2+(0.75*x[1])

#initial guesses
n=2
x0 = np.zeros(n)
x0[0] = 0.25
x0[1] = 2

#bounds
b = (-1,5)
bnds = (b,b)

con1 = {'type': 'ineq','fun': constraint1}

cons = ([con1])

result = opt.minimize(f,x0,method='SLSQP',bounds = bnds,constraints=cons)
print(result)
```
