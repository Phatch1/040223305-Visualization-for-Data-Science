<p>รางวัลโนเบล (สวีเดน: Nobelpriset; อังกฤษ: Nobel Prize) เป็นรางวัลประจำปีระดับนานาชาติ ซึ่งจัดโดยคณะกรรมการสแกนดิเนเวีย พิจารณาผลงานวิจัยหรือความอัจฉริยะและความเชี่ยวชาญที่โดดเด่น หรือสร้างคุณประโยชน์ให้กับมนุษยชาติ ทั้งในด้านวิทยาศาสตร์และวัฒนธรรม ตามเจตจำนงของอัลเฟรด โนเบล นักเคมีชาวสวีเดน ผู้ประดิษฐ์ไดนาไมท์ โดยก่อตั้งขึ้นครั้งแรกในปี ค.ศ. 1895 แต่การมอบรางวัลในสาขาฟิสิกส์ สาขาเคมี สาขาสรีรวิทยาหรือการแพทย์ สาขาวรรณกรรม และสาขาสันติภาพ เริ่มมอบรางวัลครั้งแรกในปี ค.ศ. 1901

การมอบรางวัลโนเบลสาขาสันติภาพจัดขึ้นที่เมืองออสโล ประเทศนอร์เวย์ ส่วนสาขาอื่น ๆ จัดที่เมืองสต็อกโฮล์ม ประเทศสวีเดน ผู้ได้รับรางวัลโนเบลในแต่ละสาขานั้นถือว่าเป็นเกียรติยศสูงสุดในสาขาวิชาชีพนั้น ๆ

การมอบรางวัลโนเบลจัดขึ้นเป็นประจำทุกปี ในวันที่ 10 ธันวาคม โดยผู้พระราชทานคือ สมเด็จพระราชาธิบดีแห่งราชอาณาจักรสวีเดน ถึงแม้ว่าบางปีรางวัลบางสาขาอาจไม่มีการตัดสิน แต่มีข้อกำหนดว่าระยะการเว้นการมอบรางวัลต้องไม่เกิน 5 ปี สำหรับผู้ได้รับรางวัลจะได้รับเหรียญรางวัลโนเบล ใบประกาศเกียรติคุณ เงินรางวัลประมาณ 10 ล้านโคร์นหรือประมาณ 44 ล้านบาท</p>

#ข้อ 6.

จงเขียนคำสั่งเพื่อทำสร้าง Bar Chart แสดงอัตราส่วนผู้ได้รับรางวัล Nobel Prize ที่เป็นเกิดที่ประเทศ US ในช่วงปีทศวรรต 1900 ถึง 2020 ตามรูป

Screenshot 2566-01-20 at 05.58.35.png

```python
df['usa_born_winner'] = df['born_country_code'] == 'US'
df['decade'] = (np.floor(df["year"] / 10) * 10).astype(int)
prop_usa_winners = df.groupby("decade",as_index=False)["usa_born_winner"].mean()

sns.barplot(data=prop_usa_winners, x="decade" ,y="usa_born_winner")
```

#ข้อ 7.

จงเขียนคำสั่งเพื่อทำสร้าง Line Chart แสดงอัตราส่วนผู้ได้รับรางวัล Nobel Prize ที่เป็นเกิดที่ประเทศ US ในช่วงปีทศวรรต 1900 ถึง 2020 ตามรูป

Screenshot 2566-01-20 at 06.00.01.png

```python
sns.set_theme()
sns.lineplot(data=prop_usa_winners, x="decade" ,y="usa_born_winner")
```

#ข้อ 8.

จงเขียนคำสั่งเพื่อทำสร้าง Multiple Line Chart แสดงอัตราส่วนผู้ได้รับรางวัลประเภทต่างๆ ในช่วงปีทศวรรต 1900 ถึง 2020 ตามรูป

Screenshot 2566-01-20 at 06.26.37.png

```python
df['female_winner'] = df['gender'] == 'female'
df['decade'] = (np.floor(df["year"] / 10) * 10).astype(int)
prop_female_winners = df.groupby(["decade","category"],as_index=False)["female_winner"].mean()
prop_female_winners
sns.lineplot(data=prop_female_winners, x="decade", y="female_winner", hue="category")
```
